```python
import pandas as pd 
import numpy as np
from sklearn.utils import shuffle 
import random
#Global Variables:
degree = 2
regParams = [1]*(degree+1)
weights = [random.randrange(1,10)]*(degree+1)
bias = 0.0001
#Dataset:


mainData = pd.DataFrame(pd.read_csv("seattleWeather_1948-2017.csv"))
# mainData = mainData.sample(frac=1).reset_index(drop=True)
mainData = shuffle(mainData)
train,cv,test = np.split(mainData,[int(0.6*len(mainData)),int(0.8*len(mainData))])

#Features:
featureName = 'TMAX'
feature = train[featureName]
feature = feature/max(feature)
label = train['RAIN']
cvFeature = cv[featureName]
cvFeature = cvFeature/max(cvFeature)
cvLabel = cv['RAIN']
testFeature = test[featureName]
testFeature = testFeature/max(testFeature)
testLabel = test['RAIN']
label = label.replace([False,True],[0,1])
testLabel = testLabel.replace([False,True],[0,1])
cvLabel = cvLabel.replace([False,True],[0,1])


def sigmoid(weights,x,y):
    z = 0
    for i in range(0,len(weights)):
        z = z + (weights[i]*(x**i))
    sig = 0
    sig = 1 + np.exp(-z) 
    return 1/sig

def z(weights,x,y):
    z = 0
    for i in range(0,len(weights)):
        z = z + (weights[i]*(x**i))
    return z

def logLikelihood(weights,data):
    global featureName
    oneX = data[featureName][data['RAIN']==1]
    oneX = oneX/max(oneX)
    zeroX = data[featureName][data['RAIN']==1]
    zeroX = zeroX/max(zeroX)
    oneY = data['RAIN']==1
    zeroY = data['RAIN']==0

    return sum(-np.log(1+sigmoid(weights,zeroX,zeroY)))+sum(-z(weights,oneX,oneY)-np.log(sigmoid(weights,oneX,oneY)))

def cost(weights,data,x,y,regParams):
    totalReg = 0
    #Loop for calculating Regularisation Expression for Cost Function:
    for i in range(0,len(regParams)):
        totalReg = totalReg + (regParams[i]/(len(weights)))*(sum(weights)**2)
    
    return (np.mean((logLikelihood(weights,data) - y)**2) + totalReg)


def logDerivative(weights,data,x,y,thetaIteration):
    global featureName
    oneX = data[featureName][data['RAIN']==1]
    oneX = oneX/max(oneX)
    zeroX = data[featureName][data['RAIN']==1]
    zeroX = zeroX/max(zeroX)
    oneY = data['RAIN']==1
    zeroY = data['RAIN']==0
    return np.sum(np.sum((1+np.exp(-z(weights,zeroX,zeroY)))*(np.exp(-z(weights,zeroX,zeroY)))*(-(zeroX**thetaIteration)))+(-(zeroX**thetaIteration))+np.sum(((1+np.exp(-z(weights,oneX,oneY)))*(np.exp(-z(weights,oneX,oneY)))*(-(oneX**thetaIteration)))))

def derivative(weights,data,x,y,thetaIteration,regParams):
    totalReg = 0
    #Loop for calculating Regularisation Expression for Cost Function:
    for i in range(0,len(regParams)):
        totalReg = totalReg + (regParams[i]/(len(data)))*(sum(weights)**2)
    temp = 0
    temp = (np.mean(2*(logLikelihood(weights,data) - y)) + totalReg)
    return (temp * logDerivative(weights,data,x,y,thetaIteration))+(2/len(weights)*regParams[thetaIteration]*weights[thetaIteration])

# def difference(x1,x2):
#     diff = x1-x2
    

def gradientDescent(iterations,learningRate,weights,data,x,y,regParams):
    global bias
    weightsNew = [0]*len(weights)
    for i in range(0,iterations):
        for n in range(0,len(weights)):
            weightsNew[n] = weights[n] - learningRate*derivative(weights,data,x,y,n,regParams)
            print(weights,"-----",weightsNew)
        weights = weightsNew
    return weights

iterations = 300
lr = 0.0001
learnedThetas = (gradientDescent(iterations,lr,weights,train,feature,label,regParams))


print(learnedThetas)
```
